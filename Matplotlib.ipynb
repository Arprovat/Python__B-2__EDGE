```python
price = [ 20000,30000,40000,30000,50000,20000,40000]
food = ['mango','apple','banana','orange','grapes','kiwi','pineapple']

plt.bar(food,price, width= 0.5)
plt.title('food vs price')
plt.xlabel('food')
plt.ylabel('price')
```

```python
df_=pd.read_csv('/content/drive/My Drive/Colab Notebooks/Matplotlib/Dataset/batsman_season_record.csv')
```

```python
plt.bar(df_['batsman'],df_['2015'])
```

**multi bar chart**

```python
plt.bar(np.arange(df_.shape[0])-0.25,df_['2015'],width=0.25, label = '2015')
plt.bar(np.arange(df_.shape[0]),df_['2016'],width=0.25,label='2016')
plt.bar(np.arange(df_.shape[0])+0.25,df_['2017'],width=0.25,label='2017')
```

```python
plt.bar(np.arange(df_.shape[0])-0.25,df_['2015'],width=0.25, label = '2015')
plt.bar(np.arange(df_.shape[0]),df_['2016'],width=0.25,label='2016')
plt.bar(np.arange(df_.shape[0])+0.25,df_['2017'],width=0.25,label='2017')
plt.xticks(np.arange(df_.shape[0]),df_['batsman']) #customize x axis marker and label Xticks(privious marker ,update marker)
plt.legend()
plt.tight_layout()
plt.show()
```

**stacked bar**

```python
plt.bar(df_['batsman'],df_['2015'],label='2015')
plt.bar(df_['batsman'],df_['2016'],bottom= df_['2015'],label='2016')
plt.bar(df_['batsman'],df_["2017"],bottom = (df_['2015']+df_['2016']),label='2017')
plt.legend()
plt.show()
```

```python
plt.bar(df_['batsman'],df_['2015'],label='2015')
plt.bar(df_['batsman'],df_['2016'],bottom= df_['2015'],label='2016')
plt.bar(df_['batsman'],df_["2017"],bottom = (df_['2015']+df_['2016']),label='2017')
plt.xticks(rotation=90)# rotation roted x axis marker or label
plt.legend()
plt.show()
```
